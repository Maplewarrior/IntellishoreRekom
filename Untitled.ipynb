# Modeling

```python
split = int(len(data) * 0.8)
train_encoded = data
train_y = train_encoded['transactionLocal_VAT_beforeDiscount'][:split]
train_x = train_encoded[:split].drop(['transactionLocal_VAT_beforeDiscount', 'date_rekom', 'hour', 'starthour_rounded'], axis = 1)
val_y = train_encoded['transactionLocal_VAT_beforeDiscount'][split:]
val_x = train_encoded[split:].drop(['transactionLocal_VAT_beforeDiscount', 'date_rekom', 'hour', 'starthour_rounded'], axis = 1)
```

```python
train_encoded = train_encoded.select_dtypes(exclude=np.datetime64)
```

```python
pd.DataFrame(train_encoded.dtypes)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(train_x, train_y)
```

```python
val_pre_linreg = model.predict(val_x)
```

```python
MSE = mean_squared_error(val_pre_linreg, val_y)
r2 = r2_score(val_pre_linreg, val_y)

print("MSE: ", MSE)
print("R^2: ", r2)
```
